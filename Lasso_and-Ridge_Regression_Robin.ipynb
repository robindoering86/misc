## Relations between the Distributions
#### Relation between Bernoulli and Binomial Distribution
1. Bernoulli Distribution is a special case of Binomial Distribution with a single trial.

2. There are only two possible outcomes of a Bernoulli and Binomial distribution, namely success and failure.

3. Both Bernoulli and Binomial Distributions have independent trails.

#### Relation between Poisson and Binomial Distribution
Poisson Distribution is a limiting case of binomial distribution under the following conditions:

The number of trials is indefinitely large or $n\rightarrow\infty$.
The probability of success for each trial is same and indefinitely small or $p\rightarrow 0$.
np = , is finite.
 

#### Relation between Normal and Binomial Distribution & Normal and Poisson Distribution
Normal distribution is another limiting form of binomial distribution under the following conditions:

The number of trials is indefinitely large, $n\rightarrow\infty$.
Both p and q are not indefinitely small.
The normal distribution is also a limiting case of Poisson distribution with the parameter $\lambda \rightarrow\infty$.

 

#### Relation between Exponential and Poisson Distribution
If the times between random events follow exponential distribution with rate $\lambda$, then the total number of events in a time period of length t follows the Poisson distribution with parameter $\lambda t$.

### Precision or positive predictive value (PPV)

Precision is the fraction of true positives (predicted positives, actual positives) among all positive predictions ($TP+FP$)

$$P=\frac{TP}{TP+FP}$$

### Sensitivity, recall, hit rate, or true positive rate (TPR)

Recall is the fraction of true positives (predicted positives, actual positives) of all actual positives ($TP+FN$)

$$R=\frac{TP}{TP+FN}=1-FNR$$

### Specificity / selectivity / true negative rate (TNR)

$$S=\frac{TN}{TN+FP}=1-FPR$$

### False positive rate (FPR)

$$FPR=\frac{FP}{FP+TN}=1-TNR$$

### False negative rate (FNR)

$$FNR=\frac{FN}{FN+TP}=1-TPR$$

### Accuracy

The fraction of all correct predictions:

$$A=\frac{TP+TN}{TP+TN+FP+FN}$$

### f1-score

The f1-score is the harmonic mean of precision and recall.

$$f1 = \frac{2\cdot R\cdot P}{P+R}$$

# Lasso and Ridge Regression

Lasso and ridge regression are methods to prevent overfitting  in a simple linear regression model. They do this by reducing model complexity.

## Rigde Regression

### Description

Ridge regression reduces multi-colinearities by penalizing large weights $\omega_i$ in the linear regression model 
$$ \hat{y}=+\omega_0 x_0+\omega_1+x_1+\omega_2+x_2+\ldots+\omega_n+x_n+\beta$$
through an altered cost function:

$$J=\sum^M_{i=0}\left(y_i-\sum^p_{j=0}\omega_j x_{ij}\right)^2+\lambda\sum^p_{j=0}\omega_j^2 $$

In the limit $\lambda\rightarrow 0$, the last term vanishes and the equation for $J$ takes it original form:
$$J=\sum^M_{i=0}\left(y_i-\sum^p_{j=0}\omega_j x_{ij}\right)^2$$

Hence, in ridge regression, large weights are penalizes through the last term of the cost function.

### Code example

```python
# Imports
# Numpy and Pandas
import numpy as np
import pandas as pd

# Visualization
import matplotlib.pyplot as plt
plt.rcParams.update({'font.size': 18})

# Scikit learn
from sklearn.datasets import load_boston
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Ridge
```

```python
data = load_boston()
```

```python
data.data
```

```python
boston_df=pd.DataFrame(data.data, columns=data.feature_names)
```

```python
boston_df.info()
```

```python
boston_df['price'] = data.target
```

```python
newX = boston_df.drop('price', axis=1)
newy = boston_df['price']
```

```python
X_train, X_test, y_train, y_test = train_test_split(newX, newy, test_size=.3, random_state=3)
```

```python
# Build linear regression model
lr = LinearRegression()
lr.fit(X_train, y_train)
```

```python
# Build ridge regression model
# first with low alpha of 0.01
rr = Ridge(alpha=0.01)
rr.fit(X_train, y_train)
```

```python
# Build ridge regression model
# first with large alpha of 100
rr100 = Ridge(alpha=100)
rr100.fit(X_train, y_train)
```

```python
# Calculate train and test scores for each model
lr_train_score = lr.score(X_train, y_train)
lr_test_score = lr.score(X_test, y_test)

rr_train_score = rr.score(X_train, y_train)
rr_test_score = rr.score(X_test, y_test)

rr100_train_score = rr100.score(X_train, y_train)
rr100_test_score = rr100.score(X_test, y_test)

# Print out all the different scores (6 in total)
print('Lin. Regr. train score: {} %'.format(round(lr_train_score*100, 2)))
print('Lin. Regr. test score: {} %'.format(round(lr_test_score*100, 2)))
print('Ridge Regr. train score alpha=.01: {} %'.format(round(rr_train_score*100, 2)))
print('Ridge Regr. test score, alpha=.01: {} %'.format(round(rr_test_score*100, 2)))
print('Ridge Regr. train score alpha=100: {} %'.format(round(rr100_train_score*100, 2)))
print('Ridge Regr. test score alpha=100: {} %'.format(round(rr100_test_score*100, 2)))
```

```python
plt.plot(rr.coef_, color='g', linestyle='none', marker='*', alpha=.4, label='Ridge, Alpha=0.01')
plt.plot(rr100.coef_, color='b', linestyle='none', marker='d', alpha=.4, label='Ridge, Alpha=100')
plt.plot(lr.coef_, color='r', linestyle='none', marker='o', alpha=.4, label='Linear Regr.')
plt.xlabel('Feature index')
plt.axhline(y=0, color='k')
plt.ylabel('Coefficient weight')
plt.tight_layout()
plt.locator_params(axis='y', nbins=5)
plt.ylim((-15, 7))
plt.legend(fontsize=12)
```

## Lasso Regression

Lasso = least absolute srincage and selection operator

### Description
The difference between Lasso and Ridge Regression is in the that the penalty term in the loss function $J$, the absolute value, $\mid\omega_j\mid$, of the weights is used. Hence, the equation for the loss function is now

$$J_{Lasso}=\sum^M_{i=0}\left(y_i-\sum^p_{j=0}\omega_j x_{ij}\right)^2+\lambda\sum^p_{j=0}\mid\omega_j\mid$$

In Lasso Regression, coefficients can become zero. Hence, it can not only be helpful in avoiding overfitting, but in feature selection.

### Example Code

```python
# scikit learn imports
from sklearn.linear_model import Lasso
from sklearn.datasets import load_breast_cancer
```

```python
cancer = load_breast_cancer()
cancer_df=pd.DataFrame(cancer.data, columns=cancer.feature_names)
```

```python
X = cancer.data
y = cancer.target
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.3, random_state=3)
```

```python
# Build linear regression model
lr2 = LinearRegression()
lr2.fit(X_train, y_train)
```

```python
# Build lasso regression model
# first with low alpha of 0.01
lasso = Lasso()
lasso.fit(X_train, y_train)
```

```python
# Build lasso regression model
# first with low alpha of 0.01
lasso001 = Lasso(alpha=0.01)
lasso001.fit(X_train, y_train)
```

```python
# Build lasso regression model
# first with large alpha of 0.0001
lasso0001 = Lasso(alpha=0.0001, max_iter=1e5)
lasso0001.fit(X_train, y_train)
```

```python
# Calculate train and test scores for each model
lr_train_score = lr2.score(X_train, y_train)
lr_test_score = lr2.score(X_test, y_test)

lasso_train_score = lasso.score(X_train, y_train)
lasso_test_score = lasso.score(X_test, y_test)
coeff_used = np.sum(lasso.coef_!=0)


lasso001_train_score = lasso001.score(X_train, y_train)
lasso001_test_score = lasso001.score(X_test, y_test)
coeff_used001 = np.sum(lasso001.coef_!=0)

lasso0001_train_score = lasso0001.score(X_train, y_train)
lasso0001_test_score = lasso0001.score(X_test, y_test)
coeff_used0001 = np.sum(lasso0001.coef_!=0)


# Print out all the different scores (6 in total)
print('Lin. Regr. train score: {} %'.format(round(lr_train_score*100, 2)))
print('Lin. Regr. test score: {} %'.format(round(lr_test_score*100, 2)))
print('Lasso Regr. train score alpha=1: {} %'.format(round(lasso_train_score*100, 2)))
print('Lasso Regr. test score, alpha=1: {} %'.format(round(lasso_train_score*100, 2)))
print('# Coeff. used in Lasso Regr. test score, alpha=1: {} %', coeff_used)

print('Lasso Regr. train score alpha=.001: {} %'.format(round(lasso001_train_score*100, 2)))
print('Lasso Regr. test score, alpha=.001: {} %'.format(round(lasso001_train_score*100, 2)))
print('# Coeff. used in Lasso Regr. test score, alpha=1: {} %', coeff_used001)

print('Lasso Regr. train score alpha=.0001: {} %'.format(round(lasso0001_train_score*100, 2)))
print('Lasso Regr. test score alpha=.0001: {} %'.format(round(lasso0001_test_score*100, 2)))
print('# Coeff. used in Lasso Regr. test score, alpha=1: {} %', coeff_used0001)
```
